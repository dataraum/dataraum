# DataRaum Quickstart

This notebook shows how to use the DataRaum Python API to:
1. Run the analysis pipeline on your data
2. Explore table structure and metadata
3. Check entropy scores and data readiness
4. Evaluate contracts for specific use cases
5. Query your data with natural language

**Prerequisites:** `pip install dataraum` and set `ANTHROPIC_API_KEY` in your environment.

## 1. Initialize and Run the Pipeline

Point `Context` at an output directory. If no pipeline output exists yet, use `ctx.run()` to analyze your data.

```python
from dataraum import Context

# Point to an existing pipeline output, or an empty directory for a fresh run
ctx = Context("./pipeline_output")

# Run the pipeline on your data (skip if you already have output)
# result = ctx.run("/path/to/your/data")
# result
```

## 2. Explore Tables

After the pipeline runs, typed tables are available for inspection.

```python
ctx.tables
```

## 3. Entropy Summary

Entropy measures uncertainty across 4 layers: structural, semantic, value, and computational.
Scores range from 0.0 (certain) to 1.0 (maximum uncertainty).

The result renders as an HTML table in Jupyter.

```python
# Overall entropy summary
ctx.entropy.summary()
```

```python
# Per-table entropy (replace with one of your table names)
if ctx.tables:
    ctx.entropy.table(ctx.tables[0])
```

```python
# Per-column detail with detector evidence
# ctx.entropy.details("table_name", "column_name")
```

## 4. Contract Evaluation

Contracts define acceptable entropy thresholds for specific use cases.

| Contract | Use Case |
|----------|----------|
| `exploratory_analysis` | Data exploration (lenient) |
| `data_science` | ML feature engineering |
| `operational_analytics` | Team dashboards |
| `aggregation_safe` | SUM/AVG/COUNT queries |
| `executive_dashboard` | C-level reporting |
| `regulatory_reporting` | Compliance (strict) |

```python
# List available contracts
ctx.contracts.list()
```

```python
# Evaluate a contract — is our data aggregation-safe?
ctx.contracts.evaluate("aggregation_safe")
```

## 5. Resolution Actions

When entropy is too high, DataRaum suggests prioritized actions to improve data quality.

```python
# Get prioritized actions (optionally filter by contract)
ctx.actions(contract="aggregation_safe")
```
